### Experiment 1: Consider All

In the training set the every X-ray image was annotated by three radiologist independently. There is difference in their opinion. In first experiment we trained the NN with taking consideration of contradiction of opinions. 

```python
exp_1_train_loss = np.load("/storage/home/akansh12/Vin-ChestXR-Abnormality-detection/model/exp_1_train_loss_hist_label.npy")
exp_1_test_loss = np.load("/storage/home/akansh12/Vin-ChestXR-Abnormality-detection/model/exp_1_test_loss_hist_label.npy")
exp_1_test_auc = np.load("/storage/home/akansh12/Vin-ChestXR-Abnormality-detection/model/exp_1_test_auc_hist_label.npy")
```
